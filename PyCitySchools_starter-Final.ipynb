```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
#school_data_complete_df
```

## District Summary

```python
#Get list of columns in datset
school_data_complete_df.columns
```

```python
#Calculate total number of schools.
total_schools = len(school_data_complete_df["school_name"].unique())
#print(total_schools)
```

```python
#Calculate the total number of students
total_students = len(school_data_complete_df["student_name"])
#(total_students)
```

```python
#Calculate the total budget
total_budget = school_data_complete_df["budget"].sum()
#print(total_budget)
```

```python
#Calculate the average math score 
avg_math = school_data_complete_df['math_score'].mean()
avg_reading = school_data_complete_df['reading_score'].mean()
#print(avg_math)
#print(avg_reading)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_above70_df = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70,  ['school_name','reading_score', 'math_score', 'type', 'size', 'budget', "student_name"]]
passing_math_sum = len(math_above70_df["student_name"])
#print(passing_math_sum)
percentage_math_above70=(passing_math_sum/total_students)*100  
#print(percentage_math_above70)
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
read_above70_df =school_data_complete_df.loc[school_data_complete_df["reading_score"]>=70, ["student_name",'school_name','reading_score', 'math_score', 'type', 'size', 'budget']]
passing_read_sum=len(read_above70_df["student_name"])
#print(passing_read_sum)
percentage_read_above70 = (passing_read_sum/total_students)*100
#print(percentage_read_above70)
```

```python
#Create a df that filters only passing scores
both_pass_df = school_data_complete_df[(school_data_complete_df["reading_score"]>=70) & (school_data_complete_df["math_score"]>=70)]
both_pass_df
#Calculate the percentage of students who passed math **and** reading (% Overall Passing)
passing_both_sum=len(both_pass_df["student_name"])
percentage_both_above70 = (passing_both_sum/total_students)*100
#print(percentage_both_above70)
```

```python
district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                                "Total Students": [total_students],
                                "Total Budget": [total_budget],
                                "Average Math Score": [avg_math],
                                "Average Reading Score": [avg_reading],
                                "% Passing Math": [percentage_math_above70],
                                "% Passing Reading": [percentage_read_above70],
                                "% Passing Overall": [percentage_both_above70]
                                   })
district_summary_df
```

```python
#Create bins based on budget per student
bins = [0, 585, 630, 645, 675]
group_names = ["<$584", "$585-629","$630-644", "$645-675"]
```

```python
#Create new df with limited (only needed) columns
spending_range_df=school_summary_df[["Budget Per Student", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Passing Overall"]]
#spending_range_df
```

```python
#Apply bins and groups to new df.
spending_range_df["Spending Ranges"] =pd.cut(spending_range_df["Budget Per Student"], bins, labels=group_names, include_lowest=True)
#spending_range_df
```

```python
#Find average math score per spending range
spending_math_group = spending_range_df.groupby('Spending Ranges').mean()["Average Math Score"]
```

```python
#Find average score per spending range
spending_read_group=spending_range_df.groupby('Spending Ranges').mean()["Average Reading Score"]
#print(spending_read_group)
spending_mathpass_group=spending_range_df.groupby('Spending Ranges').mean()["% Passing Math"]
#print(spending_mathpass_group)
spending_readpass_group=spending_range_df.groupby('Spending Ranges').mean()["% Passing Reading"]
#print(spending_readpass_group)
spending_bothpass_group=spending_range_df.groupby('Spending Ranges').mean()["% Passing Overall"]
#print(spending_bothpass_group)
```

```python
#Summary DF.
spending_group_summary_df = pd.DataFrame({"Average Math Score": spending_math_group,
                                          "Average Reading Score": spending_read_group,
                                         "% Passing Math": spending_mathpass_group,
                                         "% Passing Reading": spending_readpass_group,
                                         "% Passing Overall": spending_bothpass_group})
spending_group_summary_df 
```
